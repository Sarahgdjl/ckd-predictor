```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/kidney_disease.csv')
df.columns = df.columns.str.strip()

# Quick look
print(df.shape)         # (400, 26)
print(df.dtypes)
print(df.isnull().sum())

# Fix known issues in this dataset
df.replace({'?': np.nan, '\t?': np.nan}, inplace=True)
df['classification'] = df['classification'].str.strip().replace({'ckd\t': 'ckd'})

# Visualize missing values
plt.figure(figsize=(14, 5))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing values heatmap')
plt.tight_layout()
plt.show()

# Target distribution
df['classification'].value_counts().plot(kind='bar', color=['#E24B4A','#1D9E75'])
plt.title('CKD vs Non-CKD')
plt.show()
```
